```
import numpy as np
import cv2
from matplotlib import pyplot as plt

igreja = cv2.imread("carrancas.jpg")
texto = cv2.imread("Texto.png")

alturaIgreja, larguraIgreja, _ = igreja.shape
texto = cv2.resize(texto,(larguraIgreja, alturaIgreja))

marcaAgua = cv2.addWeighted(igreja, 1,texto,0.7, 0)

marcaAgua = cv2.cvtColor(marcaAgua, cv2.COLOR_BGR2RGB)
plt.imshow(marcaAgua)
plt.show()
```
