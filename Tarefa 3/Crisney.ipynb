```
import numpy as np
import cv2

ney = cv2.imread("crisface.png")
cris = cv2.imread("neyface.png")

alturaCris, larguraCris, _ = cris.shape
ney = cv2.resize(ney, (larguraCris, alturaCris))

crisney = cv2.addWeighted(cris, 0.5, ney, 0.8, 0)

from matplotlib import pyplot as plt
crisney = cv2.cvtColor(crisney, cv2.COLOR_BGR2RGB)
plt.imshow(crisney)
plt.show()
#cv2.imshow(mescri)
```
