```
import cv2
from matplotlib import pyplot as plt

mario = plt.imread("mario.jpg")
plt.imshow(mario)
plt.show()
mario = cv2.cvtColor(mario, cv2.COLOR_BGR2RGB)
altura, largura, canais_de_cores = mario.shape

plt.imshow(mario[:,:,2], 'gray')
plt.show()

plt.imshow(mario[:,:,1], 'gray')
plt.show()

plt.imshow(mario[:,:,0], 'gray')
plt.show()

contAzul=0; contVerde=0; contVermelho=0
for y in range(0,altura):
  for x in range(0,largura):
      azul,verde,vermelho = mario.item(y,x,0), mario.item(y,x,1), mario.item(y,x,2) 
      contAzul += azul ##pega o contator e soma com a quantidade de azul
      contVerde += verde ##pega o contator e soma com a quantidade de verde
      contVermelho += vermelho #pega o contator e soma com a quantidade de vermelho

tamanho=(altura,largura)
total = (altura*largura)
print("Tamanho da imagem:",tamanho,"\nTotal de pixels:",total)

mediaAzul = contAzul/total
mediaVerde = contVerde/total
mediaVermelho = contVermelho/total

print("Media Azul:",str(mediaAzul))
print("Media Verde:",str(mediaVerde))
print("Media Vermelha:",str(mediaVermelho))

if (mediaAzul > mediaVerde and mediaAzul > mediaVermelho):
  print("\nAzul é maior")
elif (mediaVerde > mediaAzul and mediaVerde > mediaVermelho):
  print("\nVerde é maior")  
elif (mediaVermelho > mediaAzul and mediaVermelho > mediaVerde):
  print("\nVermelho é maior")
else:
  print("erro")
```
