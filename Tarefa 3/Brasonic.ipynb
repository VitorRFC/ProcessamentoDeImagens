```
import numpy as np 
import cv2 
from matplotlib import pyplot as plt 

def main(): 
  sonic = cv2.imread("sonic.jpg") 
  sonic = cv2.cvtColor(sonic, cv2.COLOR_BGR2RGB) 
  plt.imshow(sonic) 
  plt.show()
  altura, largura, canais_de_cores = sonic.shape
  
  
  for y in range(0,altura):
    for x in range(0,largura):
        if sonic[y][x][2]>50 and sonic[y][x][2]<255 and sonic[y][x][1]<200 and sonic[y][x][0]<100:
          sonic[y][x][0] =0
          sonic[y][x][1] =56
          sonic[y][x][2] =59
        elif sonic[y][x][2]<179 and sonic[y][x][1]>100 and sonic[y][x][1]<255 and sonic[y][x][0]<255:
          sonic[y][x][0] =255
          sonic[y][x][1] =223
          sonic[y][x][2] =0
        elif sonic[y][x][2]<200 and sonic[y][x][1]<160 and  sonic[y][x][0]<255:
          sonic[y][x][0] =0
          sonic[y][x][1] =39
          sonic[y][x][2] =118
        
  
  plt.imshow(sonic) 
  plt.show()    
  cv2.imwrite("Brasonic.png",sonic)
main()
```
